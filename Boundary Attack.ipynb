```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
model = keras.applications.EfficientNetB0(
    include_top=True,
    weights="imagenet",
    input_tensor=None,
    input_shape=None,
    pooling=None,
    classes=1000,
    classifier_activation="softmax")
```

```python
target_sample = cv2.resize(cv2.imread('C:\\Users\\qwopq\\Downloads\\dog.jpg'),(224,224),interpolation=cv2.INTER_LINEAR)
keras.applications.efficientnet.decode_predictions(np.array(model(target_sample[tf.newaxis])),top=1)
```

```python
attack_class = np.argmax(model(initial_sample))
target_class = np.argmax(model(target_sample))

initial_sample = initial_sample[tf.newaxis]
target_sample = target_sample[tf.newaxis]
adversarial_sample = initial_sample
epsilon = 1
delta = 0.1

step = 0
```

```python
def forward_perturbation(epsilon, prev_sample, target_sample):
    perturb = (target_sample - prev_sample).astype(np.float32)
    perturb *= epsilon
    return perturb

def orthogonal_perturbation(delta, prev_sample, target_sample):
    perturb = np.random.randn(1, 224, 224, 3)
    perturb /= np.linalg.norm(perturb, axis=(1, 2))
    
    diffs = np.linalg.norm(target_sample - prev_sample, axis=(1, 2))
    
    perturb *= delta * np.mean(diffs)

    diff = (target_sample - prev_sample)
    diff /= diffs[:,tf.newaxis,tf.newaxis]

    perturb -= (np.vdot(perturb, diff) / np.linalg.norm(diff)**2) * diff
    return perturb
```

```python
while True:
    trial_sample = adversarial_sample + forward_perturbation(epsilon, adversarial_sample, target_sample)
    if np.argmax(model(trial_sample)) != attack_class:
        epsilon *= 0.9
    else:
        adversarial_sample = trial_sample
        break

while True:
    while True:
        trial_samples = []
        for i in np.arange(10):
            trial_sample = adversarial_sample + orthogonal_perturbation(delta, adversarial_sample, target_sample)
            trial_samples.append(trial_sample)
            
        d_score = np.mean(np.argmax(model(trial_samples), axis=1) == attack_class)
        if d_score <= 0.0:
            delta *= 0.9
        else:
            if d_score < 0.3:
                delta *= 0.9
            elif d_score > 0.7:
                delta /= 0.9
            adversarial_sample = np.array(trial_samples)[np.where(np.argmax(model(trial_samples), axis=1) == attack_class)[0][0]]
            break

    while True:
        print(step)
        step+=1
        
        trial_sample = adversarial_sample + forward_perturbation(epsilon, adversarial_sample, target_sample)
        if np.argmax(model(trial_sample)) != attack_class:
            epsilon *= 0.5
        else:
            adversarial_sample = trial_sample
            epsilon /= 0.5
            break
        
    if step > 10000:
        break
```
